```python
## With Axis Labels

# Variables
x = [1,2,3]
y = [2,4,6]

# Plotting
plt.plot(x,y)

# Plot Title
plt.title('Our First Graph')

# Axes Labels
plt.xlabel('X Axis') 
plt.ylabel('Y Axis')

plt.show()
```

```python
## Change fonts on Title

# Variables
x = [1,2,3]
y = [2,4,6]

# Plotting
plt.plot(x,y)

# Plot Title
plt.title('Our First Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
# "fontdict" argument is used to change font properties for title and axes

# Axes Labels
plt.xlabel('X Axis') 
plt.ylabel('Y Axis')

plt.show()
```

```python
## Change ticks on Axes

# Variables
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Plotting
plt.plot(x,y)

# Plot Title
plt.title('Our First Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
# "fontdict" argument is used to change font properties for title and axes

# Axes Labels
plt.xlabel('X Axis') 
plt.ylabel('Y Axis')

# Axes Ticks
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.show()
```

```python
## Add Legend, Change Line and Marker Properties

# Variables
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Plotting
# Also adding labels for legend and colors for lines
# plt.plot(x,y, label = '2x', color = 'red')
# plt.plot(x,y, label = '2x', color = 'yellow') # Common colors can be written in full
# plt.plot(x,y, label = '2x', color = 'yel') # gives error
# plt.plot(x,y, label = '2x', color = 'y') # Single letter color
# plt.plot(x,y, label = '2x', color = '#ababab') # Hexadecimal color

# Change Line size (width) and marker properties
plt.plot(x,y, label = '2x', color = 'red', linewidth = 2, linestyle = '--', marker = '.', markersize = 10, markeredgecolor = 'blue')

# Plot Title
plt.title('Our First Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
# "fontdict" argument is used to change font properties for title and axes

# Axes Labels
plt.xlabel('X Axis') 
plt.ylabel('Y Axis')

# Axes Ticks
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8, 10])

# Legend
plt.legend()

plt.show()
```

```python
## Add Legend, Change Line and Marker Properties (Shorthand)

# Variables
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Plotting
# Also adding labels for legend and colors for lines
# plt.plot(x,y, label = '2x', color = 'red')
# plt.plot(x,y, label = '2x', color = 'yellow') # Common colors can be written in full
# plt.plot(x,y, label = '2x', color = 'yel') # gives error
# plt.plot(x,y, label = '2x', color = 'y') # Single letter color
# plt.plot(x,y, label = '2x', color = '#ababab') # Hexadecimal color

# Change Line size (width) and marker properties
# plt.plot(x,y, label = '2x', color = 'red', linewidth = 2, linestyle = '--', marker = '.', markersize = 10, markeredgecolor = 'blue')

## Shorthand notation
# format = '[color][marker][line]'
plt.plot(x,y, 'b^-.', label = '2x')
# Check documentation for more options

# Plot Title
plt.title('Our First Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
# "fontdict" argument is used to change font properties for title and axes

# Axes Labels
plt.xlabel('X Axis') 
plt.ylabel('Y Axis')

# Axes Ticks
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8, 10])

# Legend
plt.legend()

plt.show()
```

```python
## Add Second Line Plot

# Variables
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Plotting
# Also adding labels for legend and colors for lines
# plt.plot(x,y, label = '2x', color = 'red')
# plt.plot(x,y, label = '2x', color = 'yellow') # Common colors can be written in full
# plt.plot(x,y, label = '2x', color = 'yel') # gives error
# plt.plot(x,y, label = '2x', color = 'y') # Single letter color
# plt.plot(x,y, label = '2x', color = '#ababab') # Hexadecimal color

# Change Line size (width) and marker properties
# plt.plot(x,y, label = '2x', color = 'red', linewidth = 2, linestyle = '--', marker = '.', markersize = 10, markeredgecolor = 'blue')

## Shorthand notation
# format = '[color][marker][line]'
plt.plot(x,y, 'b^-.', label = '2x')
# Check documentation for more options

## Line Number Two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2, x2**2, 'r', label = 'X^2')

# Plot Title
plt.title('Our First Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
# "fontdict" argument is used to change font properties for title and axes

# Axes Labels
plt.xlabel('X Axis') 
plt.ylabel('Y Axis')

# Axes Ticks
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8, 10])

# Legend
plt.legend()

plt.show()
```

```python
## Project Line 2 with dotted lines (See Figure)

# Variables
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Plotting
# Also adding labels for legend and colors for lines
# plt.plot(x,y, label = '2x', color = 'red')
# plt.plot(x,y, label = '2x', color = 'yellow') # Common colors can be written in full
# plt.plot(x,y, label = '2x', color = 'yel') # gives error
# plt.plot(x,y, label = '2x', color = 'y') # Single letter color
# plt.plot(x,y, label = '2x', color = '#ababab') # Hexadecimal color

# Change Line size (width) and marker properties
# plt.plot(x,y, label = '2x', color = 'red', linewidth = 2, linestyle = '--', marker = '.', markersize = 10, markeredgecolor = 'blue')

## Shorthand notation
# format = '[color][marker][line]'
plt.plot(x,y, 'b^-.', label = '2x')
# Check documentation for more options

## Line Number Two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label = 'X^2') # Solid Line
plt.plot(x2[4:], x2[4:]**2, 'r--') # Dotted Line ("Projection")

# Plot Title
plt.title('Our First Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
# "fontdict" argument is used to change font properties for title and axes

# Axes Labels
plt.xlabel('X Axis') 
plt.ylabel('Y Axis')

# Axes Ticks
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8, 10])

# Legend
plt.legend()

plt.show()
```

### Cleaner Version

```python
## LINE PLOTS

# Variables
x = [0,1,2,3,4]
y = [0,2,4,6,8]

## Plotting

# Resize the graph ( dpi specifies pixels per inch. When saving probably use 300 dpi if possible.)
plt.figure(figsize = (5,3), dpi = 100)

# Line 1

## Keyword Argument Notation
# plt.plot(x,y, label = '2x', color = 'red', linewidth = 2, linestyle = '--', marker = '.', markersize = 10, markeredgecolor = 'blue')

## Shorthand notation
# format = '[color][marker][line]'
plt.plot(x,y, 'b^-.', label = '2x') # Check documentation for more options

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label = 'X^2') # Solid Line

# Plot remaining part of graph as dashed line
plt.plot(x2[4:], x2[4:]**2, 'r--') # Dotted Line ("Projection")

# Plot Title ( Sepcify font parameters with fontdict)
plt.title('Our First Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})

# Axes Labels
plt.xlabel('X Axis') 
plt.ylabel('Y Axis')

# Axes Tickmarks ( scale of your graph)
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8, 10])

# Legend
plt.legend()

# Save Figure (dpi 300 is good when saving so the graph has high resolution)
plt.savefig('myowngraph.png', dpi = 300)

# Show plot
plt.show()
```

### Bar Chart

```python
## BAR CHARTS

# Variables
labels = ['A', 'B', 'C']
values = [1,4,2]

# Plot
bars = plt.bar(labels, values)

## Set Bar Fill Patterns
# Set Bar Patterns in Loop
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# # Set Bar Fills mannually
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

# Set Figure Size
plt.figure(figsize=(6,4))



# Show Figure
plt.show()
```
