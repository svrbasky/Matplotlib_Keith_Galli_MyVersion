````python
# Look at data
fifa.Weight
````

Counting pounds would make it hard to check weight ranges. So we strip off the lbs

````python
fifa.Weight[0]
````

````python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
````

````python
fifa.Weight
````

````python
fifa.Weight[0]
````

### Create Bins

````python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]
````

````python
medium_heavy
````

````python
heavy
````

### Plotting

````python
# Set weights
weights = [light, light_medium, medium, medium_heavy, heavy]

# Set labels
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
````

````python
# Plotting
plt.pie(weights, labels = labels)

plt.show()
````

### Change Plot Style

````python
# Set plot style
plt.style.use('ggplot')

# Set weights
weights = [light, light_medium, medium, medium_heavy, heavy]

# Set labels
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']

# Plotting
plt.pie(weights, labels = labels)

plt.show()
````

````python
# Add autopercentage
# Set plot style
plt.style.use('ggplot')

# Set weights
weights = [light, light_medium, medium, medium_heavy, heavy]

# Set labels
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']

# Plotting
plt.pie(weights, labels = labels, autopct = '%.2f %%')

plt.show()
````

The numbers are overlapping.

### Solution to overlaping labels

````python
plt.figure(figsize = (12,8))
# Add autopercentage.
# Set plot style
plt.style.use('ggplot')

# Set weights
weights = [light, light_medium, medium, medium_heavy, heavy]

# Set labels
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']

# Plotting
plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance = 0.8)# Specify percentage distances

plt.show()
````

### Explode Pie Chart

````python
plt.figure(figsize = (12,8))

# Set plot style
plt.style.use('ggplot')

# Set weights
weights = [light, light_medium, medium, medium_heavy, heavy]

# Set labels
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.1,.1,.1,.1,.1) # Explode equally

# Plotting
# Add autopercentage.
# Specify percentage distances
plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance = 0.8, explode = explode)


plt.show()
````

````python
plt.figure(figsize = (12,8))

# Set plot style
plt.style.use('ggplot')

# Set weights
weights = [light, light_medium, medium, medium_heavy, heavy]

# Set labels
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4,.2,.0,.1,.4) # Explode the smaller pieces a bit more

# Plotting
# Add autopercentage.
# Specify percentage distances
plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance = 0.8, explode = explode)


plt.show()
````

### Add title

````python
plt.figure(figsize = (12,8))

# Set plot style
plt.style.use('ggplot')

# Set weights
weights = [light, light_medium, medium, medium_heavy, heavy]

# Set labels
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4,.2,.0,.1,.4) # Explode the smaller pieces a bit more

# Plotting
# Add autopercentage.
# Specify percentage distances
plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance = 0.8, explode = explode)

plt.title('Weight Distribution of FIFA Players (in lbs)')


plt.show()
````

### Beautify the Graph

We are going to embellish the graph by highlighting the boxes and the medians. Also, we will change the color of the graph.

````python
plt.style.use('default') # reset plot style

plt.figure(figsize = (8,12))

# set labels
labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']

boxes = plt.boxplot([barcelona, madrid, revs], labels = labels)

for box in boxes['boxes']:
    # Set edge color
    box.set(color='#4286f4', linewidth=2)
    
#     # Change Fill Color
#     box.set_facecolor('#e0e0e0')

# Title and plot
plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
````

````python
plt.style.use('default') # reset plot style

plt.figure(figsize = (8,12))

# set labels
labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']

boxes = plt.boxplot([barcelona, madrid, revs], 
                    labels = labels,
                    patch_artist = True)# Set patch_artist to True to set facecolor

for box in boxes['boxes']:
    # Set edge color
    box.set(color='#4286f4', linewidth=2)
    
    # Change Fill Color
    box.set(facecolor = '#e0e0e0')

# Title and plot
plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
````

#### Set facecolor will not work unless you set ```patch_artist = True```

### Another way to embellish the plot

````python
plt.style.use('default') # reset plot style

plt.figure(figsize = (8,12))

# set labels
labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']

boxes = plt.boxplot([barcelona, madrid, revs], 
                    labels = labels, 
                    patch_artist = True, # Set patch_artist to True to set facecolor
                    medianprops={'linewidth':2}) # Set median linewidth same as box's linewidth

for box in boxes['boxes']:
    # Set edge color
    box.set(color='#4286f4', linewidth=2)
    
    # Change Fill Color
    box.set(facecolor = '#e0e0e0')

# Title and plot
plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
````
